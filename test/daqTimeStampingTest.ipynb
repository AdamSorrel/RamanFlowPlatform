```python
import time
from datetime import datetime
import numpy as np
import pandas as pd
```

```python
arr1 = np.random.rand(950)
arr2 = np.random.rand(950)
arr3 = np.random.rand(950)

dataOut = np.array([arr1, arr2, arr3])

newDataDf = pd.DataFrame(dataOut).transpose()
```

```python
newDataDf
```

```python
dataLabel = time.time_ns()/1e9
print(dataLabel)
```

```python
start = pd.Timestamp(dataLabel, unit="s")
start
```

```python
dataLen = newDataDf.shape[0]
dataLen
```

```python
samplingRate = 10000
end = start + pd.Timedelta(dataLen*1/samplingRate, unit="seconds")
```

```python
end
```

```python
end  - start
```

```python
timeIndex = pd.date_range(start=start, end=end, periods=dataLen)
```

```python
timeIndex[-1] - timeIndex[0] 
```

```python
newDataDf.index = pd.DatetimeIndex(timeIndex) 
```
