```python
from pyAndorSpectrograph.spectrograph import ATSpectrograph
from pyAndorSDK2 import atmcd, atmcd_codes, atmcd_errors
```

```python
sdk = atmcd()  # Load the atmcd library
print(f"[initialize_detector] sdk : {sdk}")
codes = atmcd_codes

ret = sdk.Initialize("")  # Initialize camera
print("[initialize_detector] Function Initialize returned {}".format(ret))
```

```python
print("Basic Example")

print("Initialising Spectrograph")
spc = ATSpectrograph() #Load the ATSpectrograph library
print("Spectrograph class loaded")
shm = spc.Initialize("")
print("Function Initialize returned {}".format(shm))

if shm == 20202:
    print("Spectrograph initialized successfully")
```

```python
(ret, devices) = spc.GetNumberDevices()
print("Function GetNumberDevices returned {}".format(spc.GetFunctionReturnDescription(ret, 64)[1]))
print("\tNumber of devices: {}".format(devices))    
```

```python
for index in range(devices):
    (ret, serial) = spc.GetSerialNumber(index, 64)
    print("Function GetSerialNumber returned {}".format(spc.GetFunctionReturnDescription(ret, 64)[1]))
    print("\tSerial No: {}".format(serial))
    
    (ret, FocalLength, AngularDeviation, FocalTilt) = spc.EepromGetOpticalParams(index)
    print("Function EepromGetOpticalParams {}".format(spc.GetFunctionReturnDescription(ret, 64)[1]))
    print("\tFocal Length: {}".format(FocalLength))         
    print("\tAngular Deviation: {}".format(AngularDeviation))
    print("\tFocal Tilt: {}".format(FocalTilt))
```

```python
shm = spc.SetGrating(0, 1)
print(f"Function SetGrating returned {spc.GetFunctionReturnDescription(shm, 64)[1]}")
```

```python
(shm, grat) = spc.GetGrating(0)
print(f"Function GetGrating returned: {grat} Grat")
```

```python
shm = spc.SetWavelength(0, 700)
print(f"Function SetWavelength returned: {spc.GetFunctionReturnDescription(shm, 64)[1]}")
```

```python
(shm, wave) = spc.GetWavelength(0)
print(f"Function GetWavelength returned: {spc.GetFunctionReturnDescription(shm, 64)[1]} Wavelength: {wave}")
```
